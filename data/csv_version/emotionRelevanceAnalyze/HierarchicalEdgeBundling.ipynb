```r
# Libraries
library(ggraph)
library(igraph)
library(tidyverse)
```

```r
# create a dataframe with connection between leaves (individuals)
all_leaves <- union(data.frame(name =read.csv(file = './emotionRelevanceWeight-dev-threadhold3-excludeOne.csv')[, 2] ), data.frame(name =read.csv(file = './emotionRelevanceWeight-dev-threadhold3-excludeOne.csv')[, 3] ))
connect <- read.csv(file = './emotionRelevanceWeight-dev-threadhold3-excludeOne.csv')[, c(2, 3)]
mygraph <- graph_from_data_frame(connect, vertices=all_leaves)

# The connection object must refer to the ids of the leaves:
from <- match( connect$node0, all_leaves$name)
to <- match( connect$node1, all_leaves$name)

ggraph(mygraph, layout = 'dendrogram', circular = FALSE) + 
  geom_edge_link() +
  theme_void()

plot
ggraph(mygraph, layout = 'dendrogram', circular = TRUE) + 
  geom_conn_bundle(data = get_con(from = from, to = to), alpha=1, colour="skyblue", tension = 0.3, width = 0.8) + 
  geom_node_point(size = .5, aes(filter = leaf, x = x*1, y=y*1)) +
  theme_void()
```
