# Start modification

## emotion

```python
# emotion col analyze
# emotion部分的脏数据
# 0. 无效emotion
# 0.1 和sentence一致
dataDevPlutchik = dataDevPlutchik[[x.strip('[').strip(']').strip('\"').strip().replace(".", "") + "." 
                                   for x in dataDevPlutchik['emotion']] != dataDevPlutchik['sentence']]
# 0.2 like something...
dataDevPlutchik = dataDevPlutchik[~ dataDevPlutchik['emotion'].str.contains("like")]

# 1. 并列
connectorList = [" and ", " but "]
for connector in connectorList:
    dataDevPlutchik['emotion'] = dataDevPlutchik['emotion'].str.lower().str.replace(connector, "\", \"") # 顺便统一大小写

# 2. 目的与原因
reasonIndentifierList = [" to", " that", " (", " about", " of", " for", " at", " in", " with", " by", " while", " on", " from", " if", " they gave"]
for reasonIndentifier in connectionList:
    dataDevPlutchik['emotion'] = dataDevPlutchik['emotion'].str.partition("\"]")[0].str.partition(reasonIndentifier)[0]+ "\"]"
    
# 3. 副词（不一定要去除其实）
decoratorList = ["very ", " a lot", " a bit", "a bit ", "little bit "]
for decorator in decoratorList:
    dataDevPlutchik['emotion'] = dataDevPlutchik['emotion'].str.replace(decorator, "")
```

```python
# remove none row
noneIndicatorList = ["none", "nothing"]
for noneIndicator in noneIndicatorList:
    dataDevPlutchik = dataDevPlutchik[~ (dataDevPlutchik['emotion'].str.contains(noneIndicator))]
```

```python
dataDevPlutchik
```

```python
# 习惯性改一下emotion和Plutchik的type, 后续好操作
dataDevPlutchik['emotion'] = [list(x.strip('[').strip(']').strip('\"').strip().split(', ')) for x in dataDevPlutchik['emotion']]
dataDevPlutchik['plutchik'] = [list(x.strip('[').strip(']').strip('\"').strip().split(', ')) for x in dataDevPlutchik['plutchik']]
dataDevPlutchik
```

```python
# 杂项清除，注意未去typo
listTotal = dataDevPlutchik['emotion']
print("curious data still...")
for index, value in listTotal.items():
    if (len(value) == 1 and value[0].count(" ") > 1):
        dataDevPlutchik = dataDevPlutchik.drop(index = index)
        print("going to remove: ", value, " @",  index)
```
